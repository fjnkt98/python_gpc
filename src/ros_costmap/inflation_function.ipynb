```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# const parameters
lethal_cost = 255
inscribed_cost = 254
circumscribed_cost = 128
```

```python
inscribed_radius = 0.1

inflation_radius = 0.55
cost_scaling_factor = 15.0
```

```python
def cost_function(x):
    return (inscribed_cost - 1) * math.exp(-cost_scaling_factor * (x - inscribed_radius))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,
                    xlabel="Distance from closest obstacle cell[m]",
                    ylabel="Cell cost",
                    )

xs = np.arange(inscribed_radius, inflation_radius, 0.01)
ys = [cost_function(x) for x in xs]

ax.plot([0, inscribed_radius], [inscribed_cost, inscribed_cost])
ax.plot(xs, ys)
plt.show()
```
