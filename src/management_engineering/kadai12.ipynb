```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def cost_function(x):
    if x <= 2304:
        return 173 * x / 60 + 5 * math.ceil(x / 10) + 441.5
    else:
        return 367 * x / 120 + 5 * math.ceil(x / 10) + 441.5 - 403.2

def profit_function(x):
    return 3*x
```

```python
xs = np.arange(0, 3000)

profit = [profit_function(x) for x in xs]
cost = [cost_function(x) for x in xs]

fig = plt.figure()
ax = fig.add_subplot(111,
                    xlabel="Production [t]",
                    ylabel="Money [10^4 Yen]",
                    )
ax.plot(xs, profit, label="Profit")
ax.plot(xs, cost, label="Cost")
ax.vlines(2304, 0, 11000, colors="black", linestyle="dotted")
plt.legend(loc="best")
plt.show()
fig.savefig("plot.pdf")
```
