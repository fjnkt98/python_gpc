```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def profit(tw, to):
    return 3.6*(tw + to)
```

```python
def cost(tw, to):
    return 3.46*tw + 3.67*to + 5*math.ceil(0.12*(tw + to)) + 261.5
```

```python
tw = np.arange(0, 1920 + 1)

p = [profit(t, 0) for t in tw]
c = [cost(t, 0) for t in tw]

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot(tw, p, label="Profit")
ax1.plot(tw, c, label="Cost")
ax1.set_xlabel("Time [h]")
ax1.set_ylabel("Money [10^4 Yen]")
plt.legend(loc='best')
plt.show()

fig1.savefig("plot1.pdf")
```
